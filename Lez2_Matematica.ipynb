# Matematica nei videogiochi 3d

### Elementi fondamentali
Un punto rappresenta una posizione nello spazio 3d.
Un vettore rappresenta un cambiamento di posizione tra punti. Il vettore non ha il concetto di posizione.

Entrambi sono una tripla di valorei
$$q \in R^3$$
$$q = 
\begin{pmatrix} 
x \\ 
y \\
z \\
\end{pmatrix}
$$

Un altro elemento importante della nostra algebra è lo scalare $\lambda$

### Operazioni basilari

**Moltiplicazione scalare per vettore (scaling)**:
Lo spostamento del vettore viene moltiplicato per lo scalare.

$$ \lambda\begin{pmatrix} 
x \\ 
y \\
z \\
\end{pmatrix} = \begin{pmatrix} 
\lambda x \\ 
\lambda y \\
\lambda z \\
\end{pmatrix}
$$

**Somma di vettori**:
Lo spostamento dei vettori viene sommato, avvengono entrambi gli spostamenti. (Regola del parallelogramma o del puntacoda). Si può anche fare la sottrazione (a-b, Il vettore risultante partirà dalla destinazione di a e la sua destinazione sarà b)

$$
\begin{pmatrix} 
x_1 \\ 
y_1 \\
z_1 \\
\end{pmatrix}
+
\begin{pmatrix} 
x_2 \\ 
y_2 \\
z_2 \\
\end{pmatrix}
=
\begin{pmatrix} 
x_1 + x_2 \\ 
y_1 + y_2\\
z_1 + z_2\\
\end{pmatrix}
$$

**Queste due operazioni sono lineari**: $\lambda (\vec{v} + \vec{w}) = \lambda\vec{v} + \lambda\vec{w}$

**Sottrazione tra punti**: trova lo spostamento tra due punti (arrivo - partenza). $\vec{v} = q - p$

**Somma punto vettore**: trova la destinazione dato una partenza e uno spostamento. $q = p + \vec{v}$

**Lunghezza di un vettore**: Lunghezza geometrica del vettore, si calcola con il teorema di pitagora generalizzato. $\vert \vec{v} \vert = \sqrt{x^2 + y^2 + z^2}$. Anche questa operazione è lineare $\vert\lambda \vec{v} \vert = \lambda \vert{v} \vert$. Inoltre vale la disuguaglianza triangolare: $$\vert \vec{v} + \vec{w} \vert \leq \vert \vec{v} \vert + \vert\vec{w} \vert$$
**Versors**: particolari tipi di vettori di lunghezza 1. Sono rappresentati così: $\hat{v}$. 

**Normalizzazione**: Passare da un vettore ad un Versor: $$\hat{d} = \vec{c} \cdot \frac{1}{\vert \vec{v} \vert}$$

**Moltiplicazione tra versors**: Il risultato è uno scalare. $$\hat{v} \cdot \hat{w} = \begin{pmatrix} 
x_v \\ 
y_v \\
z_v \\
\end{pmatrix}
\cdot
\begin{pmatrix} 
x_w \\ 
y_w \\
z_w \\
\end{pmatrix} = x_v x_w + y_v y_w + z_v z_w$$ Se il risultato di questa operazione è $0$ e nessuno dei due versors è $\vec{0}$ allora i due versors sono perpendicolari. Inoltre se questo valore è minore di 0 l'angolo dei vettori è minore di 90° altrimenti è maggiore. Inoltre dati due Versors e chiamato $\alpha$ l'angolo che formano: $$\hat d \cdot \hat n = cos(\alpha)$$

**Moltiplicazione tra vettori normali**: applicando il dot product a normali vettori otteniamo queste proprietà: 
$$ \vec v \cdot \vec w = cos(\alpha) \vert\vec v\vert\vert\vec w \vert $$
$$ \vec v \cdot \vec v = \vert\vec v \vert ^2 \rightarrow \vert\vec v \vert = \sqrt{\vec v \cdot \vec v}$$

**Moltiplicazione Vettore con Versor**: visto che la lunghezza del versor è sempre 1, $\vec v \cdot \hat n = \vert \vec v \vert \cdot cos(\alpha)$. Che moltiplicando un vettore per i versor degli assi possiamo estrarre le lunghezze delle singole coordinate del vettore.

$$
\vec v \cdot \begin{pmatrix} 
1 \\ 
0 \\
0 \\
\end{pmatrix} = x_v, \vec v \cdot \begin{pmatrix} 
0 \\ 
1 \\
0 \\
\end{pmatrix} = y_v, \vec v \cdot \begin{pmatrix} 
0 \\ 
0 \\
1 \\
\end{pmatrix} = z_v, 
$$

**Cross Product / Vectorial Product** Il risultato è un vettore, determinante della matrice tolta la riga della coordinata da calcolare. Il risultato è perpendicolare ad entrambi i vettori.

$$\vec v \times \vec w = \vec v \cdot \begin{pmatrix} 
y_v + z_w - (z_v + y_w) \\ 
z_v + x_w - (x_v + z_w) \\
x_v + y_w - (y_v + x_w) \\
\end{pmatrix}$$

Inoltre se i due vettori sono versor: $\vert \hat n \times \hat m \vert = sin(\alpha) $
Questa operazione è anticommutativa: $\vec v \times \vec w = -(\vec w \times \vec v)$




### Problemi comuni

**Interpolazione lineare tra due punti o vettori (lerp/mix/blend)**: trova un punto tra gli altri due con rapporto di distanza tra i due di t $r = p + t(q-p)$, funziona anche con i vettori e rappresenta una via di mezzo tra i due spostamenti.

**Calcolo del rimbalzo su un piano**: $\vec r = \vec v + k\hat n$ e $k = -2\vec v \cdot \hat n$
