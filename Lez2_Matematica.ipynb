# Matematica nei videogiochi 3d

### Elementi fondamentali
Un punto rappresenta una posizione nello spazio 3d.
Un vettore rappresenta un cambiamento di posizione tra punti. Il vettore non ha il concetto di posizione.

Entrambi sono una tripla di valorei
$$q \in R^3$$
$$q = 
\begin{pmatrix} 
x \\ 
y \\
z \\
\end{pmatrix}
$$

Un altro elemento importante della nostra algebra è lo scalare $\lambda$

### Operazioni basilari

**Moltiplicazione scalare per vettore (scaling)**:
Lo spostamento del vettore viene moltiplicato per lo scalare.

$$ \lambda\begin{pmatrix} 
x \\ 
y \\
z \\
\end{pmatrix} = \begin{pmatrix} 
\lambda x \\ 
\lambda y \\
\lambda z \\
\end{pmatrix}
$$

**Somma di vettori**:
Lo spostamento dei vettori viene sommato, avvengono entrambi gli spostamenti. (Regola del parallelogramma o del puntacoda). Si può anche fare la sottrazione (a-b, Il vettore risultante partirà dalla destinazione di a e la sua destinazione sarà b)

$$
\begin{pmatrix} 
x_1 \\ 
y_1 \\
z_1 \\
\end{pmatrix}
+
\begin{pmatrix} 
x_2 \\ 
y_2 \\
z_2 \\
\end{pmatrix}
=
\begin{pmatrix} 
x_1 + x_2 \\ 
y_1 + y_2\\
z_1 + z_2\\
\end{pmatrix}
$$

**Queste due operazioni sono lineari**: $\lambda (\vec{v} + \vec{w}) = \lambda\vec{v} + \lambda\vec{w}$

**Sottrazione tra punti**: trova lo spostamento tra due punti (arrivo - partenza). $\vec{v} = q - p$

**Somma punto vettore**: trova la destinazione dato una partenza e uno spostamento. $q = p + \vec{v}$

**Lunghezza di un vettore**: Lunghezza geometrica del vettore, si calcola con il teorema di pitagora generalizzato. $\vert \vec{v} \vert = \sqrt{x^2 + y^2 + z^2}$. Anche questa operazione è lineare $\vert\lambda \vec{v} \vert = \lambda \vert{v} \vert$. Inoltre vale la disuguaglianza triangolare: $$\vert \vec{v} + \vec{w} \vert \leq \vert \vec{v} \vert + \vert\vec{w} \vert$$
**Versors**: particolari tipi di vettori di lunghezza 1. Sono rappresentati così: $\hat{v}$. 

**Normalizzazione**: Passare da un vettore ad un Versor: $$\hat{d} = \vec{c} \cdot \frac{1}{\vert \vec{v} \vert}$$

**Moltiplicazione tra vettori**: Il risultato è uno scalare. $$\vec{v} \cdot \vec{w} = \begin{pmatrix} 
x_v \\ 
y_v \\
z_v \\
\end{pmatrix}
\cdot
\begin{pmatrix} 
x_w \\ 
y_w \\
z_w \\
\end{pmatrix} = x_v x_w + y_v y_w + z_v z_w$$ Se il risultato di questa operazione è $\vec{0}$ e nessuno dei due vettori è $\vec{0}$ allora i due vettori sono perpendicolari. Inoltre se questo valore è minore di 0 l'angolo dei vettori è minore di 90° altrimenti è maggiore. Inoltre dati due Versore chiamato $\alpha$ l'angolo che formano: $$\hat d \cdot \hat n = cos(\alpha)$$


### Problemi comuni

**Interpolazione lineare tra due punti o vettori (lerp/mix/blend)**: trova un punto tra gli altri due con rapporto di distanza tra i due di t $r = p + t(q-p)$, funziona anche con i vettori e rappresenta una via di mezzo tra i due spostamenti.
